**To fill null values with mean in Bench Column**

```python
df['Bench']=df['Bench'].fillna(value=df['Bench'].mean())
```

```python
df
```

```python
df['Bench']=df['Bench'].round(2)
df
```

```python
df['Bench'].isna().sum()
```

**Repeating the Same Step for agility and Sprint**

```python
columns_to_fill = ['Agility' , 'Sprint' ]
for column in columns_to_fill:
    df[column].fillna(value=df[column].mean(),inplace=True)
```

```python
df
```

```python
df[['Agility','Sprint']]=df[['Agility','Sprint']].round(2)
```

```python
df
```

**To drop Unnamed : 0 Column**

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

**Here i will fill null values in Draft Pick with the avaerage value.**

```python
df['Draft pick']=df['Draft pick'].fillna(value=df['Bench'].mean()).round(2)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.to_csv('NBA_Cleaned_Data.csv')
```
